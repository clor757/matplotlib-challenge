```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%config InlineBackend.figure_format ='retina'

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse = "Resources/mouse_drug_data.csv"
clinical = "Resources/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse = pd.read_csv(mouse)
clinical = pd.read_csv(clinical)
# Combine the data into a single dataset
main = pd.merge(mouse,clinical,on="Mouse ID")

# Display the data table for preview
main.head()
```

## Tumor Response to Treatment

```python
Drug_group = main.groupby(["Drug","Timepoint"])
Drug_meanVol = Drug_group.mean()
Drug_meanVol = Drug_meanVol.drop(columns="Metastatic Sites")
Drug_meanVol = Drug_meanVol.reset_index()
Drug_meanVol.head()
```

```python
Drug_meanpivot_df = Drug_meanVol.pivot(index="Timepoint",columns="Drug",values="Tumor Volume (mm3)")
Drug_meanpivot_df
```

```python
drug_meanErr_df = main.groupby(["Drug","Timepoint"]).sem().drop(columns={"Metastatic Sites","Mouse ID"})
drug_meanErr_df = drug_meanErr_df.reset_index()
drug_meanErr_df = drug_meanErr_df.pivot(index="Timepoint",columns="Drug",values="Tumor Volume (mm3)")
drug_meanErr_df
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 5
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size

colors=["b","g","r","c"]
x_axis=[0,5,10,15,20,25,30,35,40,45]
drug=["Capomulin","Infubinol", "Ketapril", "Placebo"]
markers=["^","o","s","d"]

for i in range(len(drug)):
    plt.plot(x_axis,Drug_meanpivot_df.loc[:,drug[i]],colors[i],marker=markers[i],linestyle="--")
    
for i in range(len(drug)):
    plt.errorbar(x_axis,Drug_meanpivot_df.loc[:,drug[i]],
                yerr = drug_meanErr_df.loc[:,drug[i]],
                color=colors[i])
    
plt.legend("best",labels=drug,ncol=1,fontsize="large")

plt.xlim(-2,47)
plt.ylim(33,73)

plt.hlines([35,40,45,50,55,60,65,70],xmin=-2,xmax=47,colors="grey")

plt.title("Tumor Respone to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

plt.show()
```

## Metastatic Response to Treatment

```python
ms_mean_df = main.groupby(["Drug","Timepoint"]).mean().drop(columns={"Tumor Volume (mm3)"})
ms_mean_df.head()
```

```python
ms_err_df = main.groupby(["Drug","Timepoint"]).sem().drop(columns={"Mouse ID","Tumor Volume (mm3)"})
ms_err_df.head()
```

```python
ms_mean_df = ms_mean_df.reset_index()
msmean_pivot_df = ms_mean_df.pivot(index="Timepoint",columns="Drug",values="Metastatic Sites")
msmean_pivot_df.head()
```

```python
ms_err_df = ms_err_df.reset_index()
mserr_pivot_df = ms_err_df.pivot(index="Timepoint",columns="Drug",values="Metastatic Sites")
mserr_pivot_df.head()
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size

colors=["b","g","r","c"]
x_axis=[0,5,10,15,20,25,30,35,40,45]
drug=["Capomulin","Infubinol", "Ketapril", "Placebo"]
markers=["^","o","s","d"]

for i in range(len(drug)):
    plt.plot(x_axis,msmean_pivot_df.loc[:,drug[i]],colors[i],marker=markers[i],linestyle="--")
    
for i in range(len(drug)):
    plt.errorbar(x_axis,msmean_pivot_df.loc[:,drug[i]],
                yerr = mserr_pivot_df.loc[:,drug[i]],
                color=colors[i])
    
plt.legend("best",labels=drug,ncol=1,fontsize="large")

plt.xlim(-2,47)
plt.ylim(-.2,3.7)

plt.hlines([0,0.5,1,1.5,2,2.5,3,3.5],xmin=-2,xmax=47,colors="grey")

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Met. Sites")

plt.show()
```

## Survival Rates

```python
srates_df = main.groupby(["Drug","Timepoint"]).count()
srates_df = srates_df.reset_index()
srates_df = srates_df.drop(columns={"Tumor Volume (mm3)","Metastatic Sites"})
srates_pivot = srates_df.pivot(index="Timepoint",columns="Drug",values="Mouse ID")
srates_pivot
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 8
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size

colors=["b","g","r","c"]
x_axis=[0,5,10,15,20,25,30,35,40,45]
drug=["Capomulin","Infubinol", "Ketapril", "Placebo"]
markers=["^","o","s","d"]

for i in range(len(drug)):
    plt.plot(x_axis,(srates_pivot.loc[:,drug[i]]/25)*100,colors[i],marker=markers[i],linestyle="--")

    
plt.legend("best",labels=drug,ncol=1,fontsize="large")

plt.xlim(-3,47)
plt.ylim(30,103)

plt.vlines([0,10,20,30,40],ymin=30,ymax=105,colors="grey")
plt.hlines([30,40,50,60,70,80,90,100],xmin=-5,xmax=47,colors="grey")

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

plt.show()
```

## Summary Bar Graph

```python
perdrug = ((Drug_meanpivot_df.loc[45]/Drug_meanpivot_df.loc[0])-1)*100
perdrug
```

```python
x_axis = np.arange(len(drug))

plt.style.use('ggplot')
plt.bar(x_axis,perdrug[drug],color=["g","r","r","r"],align="center")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, drug)

plt.ylim(-25, max(perdrug[drug])+5)

plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
```

## Observation

Capomulin shows the best results out of the four drugs we are looking at. With a 19.48% decrease in percent tumor volume over the 45 day treatment, Capomulin shows a positive result in comparison to Infubinol, Ketapril, Placebo.
